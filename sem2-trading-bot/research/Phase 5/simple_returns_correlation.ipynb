# Simple Mid Price Returns Correlation

Just calculate correlation of returns - nothing fancy.

```python
import pandas as pd
import numpy as np
from pathlib import Path

# ================================
# CHANGE THESE
# ================================
VERSION = "v7"
ROUND = "round_1"
INSTANCE = "instance_1"

# Load data
csv_file = Path(f"../raw_data/{VERSION}/{ROUND}/{INSTANCE}.csv")
df = pd.read_csv(csv_file)

print(f"Loaded {len(df)} rows from {VERSION} {ROUND} {INSTANCE}")
```

## Calculate Returns Correlation

```python
# Calculate returns
df['returns'] = df['mid_price'].pct_change()
returns = df['returns'].dropna()

print(f"Returns calculated: {len(returns)} observations")
print(f"Mean return: {returns.mean():.6f}")
print(f"Std dev: {returns.std():.6f}")

# Calculate correlations with lagged returns
print("\n📊 CORRELATIONS:")
print("=" * 30)

for lag in [1, 2, 3, 4, 5, 10, 20]:
    if len(returns) > lag:
        corr = returns.autocorr(lag=lag)
        print(f"Lag {lag:2d}: {corr:7.4f}")

# Find the strongest correlation
correlations = {}
for lag in range(1, min(21, len(returns))):
    correlations[lag] = returns.autocorr(lag=lag)

strongest_lag = max(correlations.keys(), key=lambda k: abs(correlations[k]))
strongest_corr = correlations[strongest_lag]

print(f"\n🔥 Strongest: Lag {strongest_lag} = {strongest_corr:.4f}")

if abs(strongest_corr) > 0.1:
    print("⚠️  Significant correlation detected!")
elif abs(strongest_corr) > 0.05:
    print("⚡ Weak correlation")
else:
    print("✅ No strong correlation (random walk)")
```
